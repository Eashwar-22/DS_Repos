----
### MULTI VARIABLE REGRESSION
----

````python
data.head()
````

````python
X = data.drop(columns='Price')
y = price_log

X_train,X_test,y_train,y_test = train_test_split(X,
                                                 y,
                                                 test_size=0.2,
                                                 random_state=10,)

reg = LinearRegression()
reg.fit(X_train,y_train)

print("Train score : ",reg.score(X_train,y_train))
print("Test score : ",reg.score(X_test,y_test))

print("Intercept : ",reg.intercept_)
pd.DataFrame(data=reg.coef_.round(3),index=X_train.columns,columns=['Coefficient'])
````

----
### MODEL SIMPLIFICATION
----

--------
#### P Value Significance

$ p <= 0.05 $; <br>```Statistically significant``` <br>
$ p > 0.05 $; <br>```Insignificant```

-------


````python
X_include_intercept = sm.add_constant(X_train)
model = sm.OLS(y_train,X_include_intercept)
results = model.fit()
org =pd.DataFrame({'coef' : results.params.round(3),'p_value' : results.pvalues.round(3)})
org_temp = org.copy()
org_temp['significant'] = org_temp['p_value']<=0.05
org_temp
````

--------
#### Variable Inflation Factor

$ FeatureX = \alpha_0 + \alpha_1 Feature 1 + \alpha_2 Feature 2  + ... $
<br>
<br>
$ VIF_{FeatureX} = \frac{1}{{1-R_{FeatureX}^2}} $

-------


````python
def find_vif(x):
    return round(variance_inflation_factor(exog=np.array(X_include_intercept),
                                     exog_idx=x),2)

vif_df = pd.DataFrame(index=X_include_intercept.columns).reset_index().reset_index()
vif_df.columns=['col_no','col']
vif_df['VIF'] = vif_df['col_no'].apply(lambda x : find_vif(x))
vif_df.drop(columns='col_no')
````

```VIF < 10 for all coefficients so multicollinearity is not a concern in this case.```<br>
```p > 0.05   in   INDUS, AGE  --> Their insignificance could be justified using BIC```

#### Baysian Information Criterion

_Model including AGE & INDUS_

````python
X_include_intercept = sm.add_constant(X_train)
model = sm.OLS(y_train,X_include_intercept)
results = model.fit()
print("BIC : ",results.bic)
print("R^2 : ",results.rsquared)
````

_Model excluding AGE & INDUS_

````python
X_include_intercept = sm.add_constant(X_train)
X_include_intercept.drop(columns=['AGE','INDUS'],inplace=True)
model = sm.OLS(y_train,X_include_intercept)
results = model.fit()
print("BIC : ",results.bic)
print("R^2 : ",results.rsquared)
````

```Lesser the BIC, better the model```

````python
org2 =pd.DataFrame({'coef_reduced' : results.params.round(3),'p_value_reduced' : results.pvalues.round(3)})
org.join(org2)
````

----
### RESIDUAL ANALYSIS
----

**Modified Model - removing INDUS, AGE from Training set.**

````python
# Regression model for modified data
X = data.drop(columns=['Price','AGE','INDUS'])
y = price_log  # Log of Price

X_train,X_test,y_train,y_test = train_test_split(X,
                                                 y,
                                                 test_size=0.2,
                                                 random_state=10,)
reg = LinearRegression()
reg.fit(X_train,y_train)

print("Train score : ",round(reg.score(X_train,y_train),2))
print("Test score : ",round(reg.score(X_test,y_test),2))

# New OLS
X_include_intercept = sm.add_constant(X_train)
model = sm.OLS(y_train,X_include_intercept)
results = model.fit()
org =pd.DataFrame({'coef' : results.params.round(3),'p_value' : results.pvalues.round(3)})

# Residual
residuals = results.resid # y_train - predicted values(fitted values)
y_pred = results.fittedvalues

corr = round(y_train.corr(y_pred),2)
print("y_train vs y_predicted Correlation : ",corr)

# Plot Actual vs Predicted values
plt.figure(figsize=(16,7))

# For log(price)
plt.subplot(121)
plt.xlabel('Actual log prices $y_{log}$',fontsize=16)
plt.ylabel('Predicted log prices $\hat y_{log}$',fontsize=16)
plt.title('$y_{log}$  vs $\hat y_{log}$ with Correlation = '+str(corr),fontsize=16)
sns.scatterplot(x=y_train,y=y_pred)
sns.lineplot(x=y_train,y=y_train,color='red')

# For actual price [taking exponential of log(price)]
plt.subplot(122)
plt.xlabel('Actual prices in 1000 dollars $y$',fontsize=16)
plt.ylabel('Predicted prices in 1000 dollars $\hat y$',fontsize=16)
plt.title('$y$  vs $\hat y$ with Correlation = '+str(corr),fontsize=16)
sns.scatterplot(x=np.e**y_train,y=np.e**y_pred)
sns.lineplot(x=np.e**y_train,y=np.e**y_train,color='red')

plt.show()

plt.figure(figsize=(16,7))
plt.xlabel('Predicted log prices $\hat y_{log}$',fontsize=16)
plt.ylabel('Residual',fontsize=16)
plt.title('Predicted Values vs Residual',fontsize=16)
sns.scatterplot(x=y_pred,y=residuals,color='green')

plt.show()
````

```A pattern is observed in Houses costing around 50,000 dollars. Apart from them, the residuals appear to be randomly distributed.```

**Univariate Distribution of the Residuals**

````python
res_mean = round(residuals.mean(),3)
res_skew = round(residuals.skew(),3)

plt.figure(figsize=(8,6))
plt.title(f'Residual Distribution [Skew ~ {res_skew}, Mean ~ {res_mean}]',fontsize=16)
plt.xlabel('Residual',fontsize=16)
sns.distplot(x=residuals,color='purple')

plt.show()
````

```Residual curve is almost normally distributed with slightly longer tails at both ends.```

----
### MAKING PREDICTIONS
----

````python
reduced_log_mse = round(results.mse_resid,3)
reduced_log_rmse  = round(np.sqrt(reduced_log_mse),3)
reduced_log_rsquared = round(results.rsquared,3)
print("Mean squared error : ",reduced_log_mse)
print("Root Mean squared error : ",reduced_log_rmse)
print("R squared : ",reduced_log_rsquared)
````

---
**Prediction Intervals**

Setting a price range for predicted values. <br>

```Upper limit -> prediction + 2 S.D (95%)```<br>
```Lower limit -> prediction - 2 S.D (95%)```<br>

(S.D set using RMSE of Residuals ; refer to above distribution plot)

---

````python
# Prediction intervals for log(price)

pred_interval_log = pd.DataFrame({'y_train':y_train.round(3),'y_pred':y_pred.round(3)})
pred_interval_log['Lower bound (log price)'] = (pred_interval_log['y_pred'] - 2*reduced_log_rmse).round(3)
pred_interval_log['Upper bound (log price)'] = (pred_interval_log['y_pred'] + 2*reduced_log_rmse).round(3)
pred_interval_log
````

````python
# Prediction intervals for price (in 1000 dollars)

pred_interval = pred_interval_log.copy()
pred_interval['Actual price'] = pred_interval['y_train'].apply(lambda x : round(np.e**(x),3))
pred_interval['Predicted price'] = pred_interval['y_pred'].apply(lambda x : round(np.e**(x),3))
pred_interval['Lower bound'] = pred_interval['Lower bound (log price)'].apply(lambda x : round(np.e**(x),3))
pred_interval['Upper bound'] = pred_interval['Upper bound (log price)'].apply(lambda x : round(np.e**(x),3))
pred_interval = pred_interval[['Actual price', 'Predicted price','Lower bound', 'Upper bound']]

print("Prices in $1000")
pred_interval
````
